```python
import json
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay,classification_report
import matplotlib.pyplot as plt
from rich import print
```

```python
path = f'stats/stats_densenet_NOFAM.json'

data =json.load(open(path))
data.keys()
```

```python
data["train_true_classes"] = [int(x) for x in data["test_true_classes"]]
data["train_pred_classes"] = [int(x) for x in data["test_pred_classes"]]
```

```python
cmat = confusion_matrix(data["train_true_classes"], data["train_pred_classes"])
ConfusionMatrixDisplay(confusion_matrix=cmat,display_labels=["gpd","grda","lpd","lrda","other","seizure"]).plot(cmap=plt.cm.cividis)
plt.title("Confusion Matrix- DenseNet without FAM: Epoch 18")
plt.show()
```

```python
print(classification_report(data["train_true_classes"], data["train_pred_classes"], target_names=["gpd","grda","lpd","lrda","other","seizure"]))
```
